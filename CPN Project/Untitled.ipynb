```python
import numpy as np
V = 1
k0 = 0.0001
Ea = 60000
R = 8.314
T = 866.85+273.15
K = k0*np.exp(-Ea/(R*T))
CAos = 258.83
CBos = 776.5
CAs = 82.45
CBs = 247.36
CCs = 352.76
Fs = 1.26
print(K)
```

```python

import sympy as sp
```

```python
a11 = -(Fs + V*K*CBs**3)/V
a12 = -3*K*CAs*(CBs**2)
a13 = 0

a21 = -3*K*V*CBs**3
a22 =-(Fs + 9*V*K*CAs*(CBs**2))/V
a23 = 0

a31 = (2*V*K*CBs**3)/V
a32 = 6*K*CAs*CBs**2
a33 = -Fs/V

b11 = (CAos - CAs)/V
b21 = (CBos - CBs)/V
b31 = (-CCs)/V

c11 = Fs/V
c21 = 0
c31 = 0
```

```python
A = sp.Matrix([[a11, a12, a13],[a21, a22, a23], [a31, a32, a33]])
B = sp.Matrix([[b11],[b21],[b31]])
E = sp.Matrix([[c11],[c21],[c31]])
C = sp.Matrix([[0, 0, 1]])
I = sp.eye(A.shape[0])

Cc, F, s = sp.symbols('Cc, F, s')
Gp = C*(s*I-A).inv()*B
Gp = sp.nsimplify(Gp[0,0]).simplify()
display(sp.Eq(Cc/F,Gp))
```

```python
s, F, P = sp.symbols('s, F, P')

tau = 10
Kp = 1/12
zeta = 3
P_initial = 9
del_P = 4
Ccv = 0.09622
xs = Kp*P_initial

rho = 8.81609
SG = rho/1000
F_P = (Ccv*np.sqrt(del_P/(xs*SG))/((2)*(12)*(100*s**2+60*s+1)))
display(sp.Eq(F/P, F_P))
```

```python
ans = F_P*Gp
display(sp.Eq(Cc/P, ans))
```

```python
poles = list(sp.roots(sp.denom(ans)).keys())
zeros = list(sp.roots(sp.numer(ans)).keys())
display(poles)
display(zeros)
```
