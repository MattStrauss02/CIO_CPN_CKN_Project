#### Polytropic calculations

We will look at the polytropic calculations of the compressor to determine the volume of the compressor that is necessary to move the required amount of fluid as well as compressing the fluid to a certain pressure. The rpm of the crank also needs the calculated. Below shows the P vs V graph of the process

```python
#Values from DWSIM
PA = 30*100 #KPa
PC = 70*100 #KPa
nAdia = 0.75 #Selected Adiabatic efficiency
nPoly= 0.76 #Polytropic efficiency
W = 982.6 #KW required
Tin = 298.15 #K
Tout = 298.15 + 34.5464 #K
cAdia = 1.432
cPoly = 1.6466
rpm = 1500
rp = (PC - PA)/PA
print(rp)
```

Formulas and method was obtained from https://youtu.be/Bpr1KiJz0B0

C = VC/(VA-VC)

nV = 1 - C((rp)**(1/nPoly) - 1)

nV = (VA-VD)/(VA-VC)

DelV = VA - VD

```python
VA = 0.72  #m3 just guessing for now
DelV = 0.7  #For Wpoly to be = W
VD = VA - DelV
print(VD)
```

```python
def sol(VC):
    ans = 1 - (VC/(VA-VC))*((rp)**(1/nPoly) - 1)-(VA-VD)/(VA-VC)
    return ans
```

```python
print('VC =',fsolve(sol, 0.01), 'm3')
```

```python
Wpoly = PA*(nPoly/(nPoly-1))*(rp**(nPoly-1/nPoly)-1)*DelV
print(Wpoly)
```

```python
Q = 0.8174
rpm = (Q/DelV)*60
print('rpm =',rpm)
```

Changing VA does not affect the power or the rpm of the compressor. We want VA to be as small as possible to save on costs for the compressor, we do however need to leave enough clearance room for the compressor to allow smooth suction and discharge of the fluid. Thus, a VA of 0.72 m3 is used which leaves us with a clearance room of 0.0137 m3. This is enough clearance room whilst also having a VA that is as low as possible.
